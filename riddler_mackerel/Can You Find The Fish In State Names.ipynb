```python
import pandas as pd
import numpy as np
```

```python
states = []
with open('states.txt', 'r') as f:
    for line in f:
        states.append(line.strip())

for i in range(50):
    states[i] = states[i].lower()

states
```

```python
words = []
with open('word.list.txt', 'r') as f:
    for line in f:
        words.append(line.strip())
    
words
```

```python
def common(word1, word2):
    letters_in_common = ''.join(set(word1) & set(word2))
    return len(letters_in_common)

# i hate capitals
print(common('Mississippi', 'mackerel'))
print(common('mississippi', 'mackerel'))
```

```python
def unique():
    data = []
    for state in states:
        unshared = []
        for word in words:
            if (common(state, word) == 0):
                unshared.append(word)
        d = {state: unshared}
        data.append(d)
    return data
uniques = unique()
```

```python
df = pd.DataFrame(uniques)
df_cleaned = pd.DataFrame()
for i in range(50):
    df_cleaned = pd.concat([df_cleaned, df.iloc[:, i].dropna().rename(index = {i : 0})], axis = 1)
df_cleaned = df_cleaned.T.rename(columns = {0 : 'unshared'})
df_cleaned

# checking if 'mackerel' works on ohio (index 34)
for i in range(50):
    if 'mackerel' in df_cleaned.iloc[i][0]:
        print(i)
```
